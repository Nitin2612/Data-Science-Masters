#Q1. What is an Exception in python? Write the difference between Exceptions and Syntax errors.

###An exception in Python is an event or an error that disrupts the normal flow of a program's execution. Exceptions are typically raised when a program encounters an error during runtime, and they provide a way to handle errors without abruptly terminating the program. Python provides a built-in mechanism to handle exceptions using try, except, else, and finally blocks.

##Types of Exceptions in Python:
- ZeroDivisionError: Occurs when attempting to divide a number by zero.
FileNotFoundError: Raised when trying to access a file that does not exist.
- ValueError: Raised when a function receives an argument of the correct type but an inappropriate value.
- IndexError: Raised when trying to access an element from a list using an invalid index.

#Q3. Which Python statements are used to catch and handle exceptions? Explain with an example.

###In Python, exceptions are caught and handled using the try, except, else, and finally statements.

##1. try Statement:
- This block contains the code that might raise an exception.
- If an exception occurs in the try block, the program will stop executing the rest of the code in the try block and jump to the corresponding except block.

##2. except Statement:
- This block catches the exception that was raised in the try block.
- You can specify the type of exception to catch (e.g., ZeroDivisionError, ValueError), or you can catch all exceptions with a general except clause.

##3. else Statement:
- This block runs if no exceptions are raised in the try block. Itâ€™s useful for code that should only run when no exceptions occur.
- The else block is optional and executes only if no error occurred in the try block.

##4. finally Statement:
- This block is always executed, regardless of whether an exception was raised or not.
- It is generally used for clean-up actions (e.g., closing files or releasing resources) that must happen regardless of whether an exception occurred.


```python
#Example:
try:
    num = int(input("Enter a number: "))
    result = 10 / num
    print(f"Result is: {result}")
except ZeroDivisionError:
    print("Error: Cannot divide by zero!")
except ValueError:
    print("Error: Invalid input! Please enter a valid number.")
else:
    print("The division was successful!")
finally:
    print("This is the end of the program.")
```

#Q4. Explain with an example:
#a. try and else
#b. finally
#c. raise

###a. try and else
- The try block allows you to write code that might raise an exception, while the else block is used to define code that should only execute if no exceptions occur in the try block. The else block is optional, and it helps in situations where you want to execute some code only when the try block runs without any errors.

###b. finally
- The finally block is always executed, regardless of whether an exception occurs or not. It's typically used for cleanup actions, such as closing files, releasing resources, or closing database connections.

###c. raise
- The raise statement is used to manually trigger an exception. It can be used to raise a specific exception with a custom message or to re-raise an exception after it has been caught in an except block.

```python
#Example: try and else
try:
    num1 = int(input("Enter a number: "))
    num2 = int(input("Enter another number: "))
    result = num1 / num2
except ZeroDivisionError:
    print("Error: Cannot divide by zero!")
except ValueError:
    print("Error: Please enter valid integers.")
else:
    print(f"The result of division is: {result}")
```

```python
#Example: finally
try:
    file = open("testfile.txt", "r")
    content = file.read()
    print(content)
except FileNotFoundError:
    print("Error: The file does not exist!")
finally:
    print("Closing the file.")
    try:
        file.close()
    except NameError:
        print("File was not opened, so no need to close it.")
```

```python
#Example: raise
def check_age(age):
    if age < 18:
        raise ValueError("Age must be 18 or older!")
    else:
        print("Access granted.")

try:
    age = int(input("Enter your age: "))
    check_age(age)
except ValueError as e:
    print(f"Error: {e}")
```

#Q5. What are Custom Exceptions in python? Why do we need Custom Exceptions? Explain with an example.

###In Python, exceptions are used to handle errors and abnormal conditions during the execution of a program. While Python provides a set of built-in exceptions, there are situations where predefined exceptions might not be sufficient for handling specific error conditions that arise in custom programs. In such cases, Custom Exceptions are created.

####A Custom Exception is a user-defined exception class that extends the base exception class (Exception) or any of its subclasses. Custom exceptions provide a way to define error conditions specific to the program and allow for better error handling and reporting.

##Why do we need Custom Exceptions?
- Specificity: Built-in exceptions like ValueError or IndexError might not provide enough information to indicate the cause of an error specific to the application. Custom exceptions allow you to create specific error messages for particular scenarios.
- Better Error Handling: Custom exceptions provide better control over how exceptions are handled, enabling more targeted error handling strategies.
- Improved Debugging: Custom exceptions allow you to include additional information (e.g., error codes, messages) that can help in debugging and tracing errors in your program.

```python
#Example:
class InsufficientBalanceError(Exception):
    def __init__(self, message="Insufficient balance for the transaction."):
        self.message = message
        super().__init__(self.message)

def withdraw_amount(account_balance, withdrawal_amount):
    if withdrawal_amount > account_balance:
        raise InsufficientBalanceError("You cannot withdraw more than the available balance.")
    else:
        account_balance -= withdrawal_amount
        return account_balance

try:
    account_balance = 500
    withdrawal_amount = 600

    account_balance = withdraw_amount(account_balance, withdrawal_amount)
    print(f"New balance: {account_balance}")
except InsufficientBalanceError as e:
    print(f"Error: {e}")
```

#Q6. Create a custom exception class. Use this class to handle an exception.

```python
class CustomException(Exception):
    def __init__(self, message):
        super().__init__(message)

def divide_numbers(a, b):
    try:
        if b == 0:
            raise CustomException("Division by zero is not allowed.")
        return a / b
    except CustomException as e:
        print(f"Error: {e}")

result = divide_numbers(10, 0)
result = divide_numbers(10, 2)
if result is not None:
    print(f"Result: {result}")
```
